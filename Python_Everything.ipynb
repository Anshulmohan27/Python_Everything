# **Variables and Types**

```python
#printing method
mystring = "hello"
myfloat = 10.0
myint = 20

if mystring == "hello":
    print("String: %s" % mystring)
if isinstance(myfloat, float) and myfloat == 10.0: #checking if number is float and equal to 10.0
    print("Float: %f" % myfloat)
if isinstance(myint, int) and myint == 20: #checking if number is integer and equal to 20
    print("Integer: %d" % myint)
```

# **Lists**
Lists are very similar to arrays. They can contain any type of variable, and they can contain as many variables as you wish. Lists can also be iterated over in a very simple manner.

```python
my_list = [] #square brackets intitialize lists
my_list.append(1)
my_list.append(2)
print(my_list)

my_list.append('1')
print(my_list)

print(my_list[0]) #first entry
print(my_list[2]) #third entry is string
```

# **Basic Operators**

```python
number = 1 + 2 * 3 / 4.0  #basic arithmetic
print(number)

remainder = 11 % 3   #remainder calculation
print(remainder)

squared = 7 ** 2   #squared
cubed = 2 ** 3
print(squared)
print(cubed)

helloworld = "hello" + " " + "world"
print(helloworld)

lotsofhellos = "hello" * 10
print(lotsofhellos)

even_numbers = [2,4,6,8]
odd_numbers = [1,3,5,7]
all_numbers = odd_numbers + even_numbers #joining two lists
print(all_numbers)

print([1,2,3] * 3)  #does not print multiplied by three individually but list multiple
```

# **String Formatting**

```python
name = "John"
print("Hello, %s!" %name)

age = 23
print("%s is %d years old." %(name, age)) #%s - String, %d - Integers, %f - Floating point numbers, %.<number of digits>f - Floating point numbers with a fixed amount of digits to the right of the dot.

mylist = [1,2,3]
print("A list: %s" % mylist)

data = ('Hello', 'John', 'Doe', 59.9)
print('%s, %s %s, your current balance is %.2f.' %data)
```

# **Basic String Operations**

```python
my_string = "Hello, world, this is a string!"

print(len(my_string))

print(my_string.index('o')) #first instance

print(my_string.count('t'))

print(my_string[3:8])

print(my_string[3:8:2])

print(my_string[::-1])

print(my_string.upper())

print(my_string.lower())

print(my_string.startswith("Hello"))

print(my_string.endswith("asdfasdfasdf"))

my_string1 = my_string.split(" ")

print(my_string1)

print(my_string[10]) #tenth index
```
