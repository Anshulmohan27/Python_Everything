# **Classes and Objects**

```python
class MyClass:
    variable = "blah"

    def function(self):
        print("This is a message inside the class.")

myobjectx = MyClass()
print(myobjectx.variable)
myobjectx.function()

myobjecty = MyClass()

myobjecty.variable = "yackity"

print(myobjecty.variable)

class NumberHolder:

   def __init__(self, number):
       self.number = number

   def returnNumber(self):
       return self.number

var = NumberHolder(7)
print(var.returnNumber()) #Prints '7'

# define the Vehicle class
class Vehicle:
    name = ""
    kind = "car"
    color = ""
    value = 100.00
    def __init__(self, name, color, value):
        self.name = name
        self.color = color
        self.value = value
    def description(self):
        desc_str = "%s is a %s %s worth $%.2f." % (self.name, self.color, self.kind, self.value)
        return desc_str

car1 = Vehicle("Fer", "red", 60000)
car2 = Vehicle("Jump", "blue", 10000)
# test code
print(car1.description())
print(car2.description())

class Mobile:

  def __init__(self, brand, color, price):
    self.brand = brand
    self.color = color
    self.price = price

  def print_mobile(self):
    pdesc = "The %s brand mobile phone is of %s color and priced at %d." %(self.brand, self.color, self.price)
    return pdesc

Moto_G = Mobile("Motorola", "Blue", 15000)
print(Moto_G.print_mobile())
```

# **Generator**

```python
import random

def lottery():
  for i in range(6):
    yield random.randint(1,40)

for random_number in lottery():
  print("The lottery number is %d" %(random_number))
```

```python
def fib(arg):
    a,b=1,1
    for i in range(arg):
        yield a
        a,b = b,a+b

for j in fib(10):
    print (j)
```

# **Map, Filter, Reduce**

```python
my_pets = ['alfred', 'tabitha', 'william', 'arla']

uppered_pets = list(map(str.upper, my_pets))

print(uppered_pets)
```

```python
circle_areas = [3.56773, 5.57668, 4.00914, 56.24241, 9.01344, 32.00013]

result = list(map(round, circle_areas, range(1, 7)))

print(result)
```

```python
my_strings = ['a', 'b', 'c', 'd', 'e']
my_numbers = [1, 2, 3, 4, 5]

results = list(zip(my_strings, my_numbers))

print(results)
```

```python
my_strings = ['a', 'b', 'c', 'd', 'e']
my_numbers = [1, 2, 3, 4, 5]

results = list(map(lambda x, y: (x, y), my_strings, my_numbers))

print(results)
```

```python
from functools import reduce

my_floats = [4.35, 6.09, 3.25, 9.77, 2.16, 8.88, 4.59]
my_names = ["olumide", "akinremi", "josiah", "temidayo", "omoseun"]
my_numbers = [4, 6, 9, 23, 5]

map_result = list(map(lambda x: round(x ** 2, 2), my_floats))
filter_result = list(filter(lambda name: len(name) <= 7, my_names))
reduce_result = reduce(lambda num1, num2: num1 * num2, my_numbers)

print(map_result)
print(filter_result)
print(reduce_result)
```
