# **String Formatting**

```python
name = "John"
print("Hello, %s!" %name)

age = 23
print("%s is %d years old." %(name, age)) #%s - String, %d - Integers, %f - Floating point numbers, %.<number of digits>f - Floating point numbers with a fixed amount of digits to the right of the dot.

mylist = [1,2,3]
print("A list: %s" % mylist)

data = ('Hello', 'John', 'Doe', 59.9)
print('%s, %s %s, your current balance is %.2f.' %data)
```

# **Basic String Operations**

# **Functions**

```python
def func():
  return "Hello, world!"

print(func())

def func_w_arg(greeting, name):
  string_1 =  greeting + ', ' + name
  return string_1

print(func_w_arg("Hello", "John"))

def sum_2num(a, b):
  return a+b

print(sum_2num(12,23))

def list_benefits():
    return ["More organized code", "More readable code", "Easier code reuse", "Allowing programmers to share and connect code together"]

def build_sentence(benefit):
    return benefit + " is a benefit of functions!"

def construct_sent():
    benefits = list_benefits()
    for some in benefits:
      print(build_sentence(some))

construct_sent()

```
