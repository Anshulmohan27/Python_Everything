```python
#Create a column categorizing packet count:
#Low (<50)
#Medium (50-150)
#High (>150)
def assign_cat(df):
    if df['packet_count'] < 50:
        return 'Low'
    elif (df['packet_count'] >= 50) & (df['packet_count'] <150):
        return 'Medium'
    elif df['packet_count'] >= 150:
        return 'High'

df['packet_category'] = df.apply(assign_cat, axis=1)
df[['packet_category', 'packet_count']]
```

```python
#Which protocol generates the most traffic overall?
df['total_traffic'] = df['dst_bytes'] + df['src_bytes']
df.groupby('protocol')['total_traffic'].size()
```
